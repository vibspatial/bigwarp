## Switch moving and fixed in Bigwarp landmarks CSV
In some cases, you may want to easily switch the moving and fixed landmarks in your landmarks.csv file. For example, you mixed them up at the beginning of your registration and only realized halfway through. In those cases, you can just change the paths in the code below and run the cell to correct your file.

```python
import csv

# Input and output file paths
input_file = r'path/to/landmarks.csv'
output_file = r'path/to/swapped_landmarks.csv'

# Open the input file for reading and the output file for writing
with open(input_file, 'r', newline='') as csvfile_in, open(output_file, 'w', newline='') as csvfile_out:
    reader = csv.reader(csvfile_in, quotechar='"')
    writer = csv.writer(csvfile_out, quotechar='"', quoting=csv.QUOTE_ALL)

    # Iterate through each row in the input CSV
    for row in reader:
        # Swap columns
        row[2], row[4] = row[4], row[2]
        row[3], row[5] = row[5], row[3]

        # Write the modified row to the output file
        writer.writerow(row)
```
